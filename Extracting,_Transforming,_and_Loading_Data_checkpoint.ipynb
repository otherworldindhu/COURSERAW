# Aggregate the `transactions` data

```python
# Aggregate data on the number of transactions and the total amount.
query4='select user_id, count(transaction_id) no_of_transactions , sum(amount_usd) tot_amt from transactions group by user_id'

agg=pd.read_sql(query4,connection_pa)
agg
```

# Merge the `device` table with the `users` table

```python
# Do a left join, as all users in the users table are of interest.
query5 = 'select u.*, d.device from users u left join device d on u.user_id=d.user_id'

users_w_device = pd.read_sql(query5, connection_pa)
users_w_device.head()
```

# Close the database connection

```python
connection_pa.close()
```

```python
users_w_device.columns, agg.columns
```

# Merge `users_w_device` with `transactions_agg`

```python
# Do a right join so users won't be lost.
# at this point, there's so sql connection.
merged=agg.merge(users_w_device,how='right', on='user_id')
merged.head()



```

```python
# Make sure number of rows is equal to users_w_devices table.

merged.shape, users_w_device.shape
```

# Identify data where `age` is greater than 150

```python
#there's no more passing query, the connection is already close

merged[merged.age>150]
```

# Drop incorrect data

```python
users_cleaned_pa=merged[merged.age<150]
users_cleaned_pa.head()
```

# Identify more potentially erroneous data

```python
# Compare age to device.
pd.crosstab(users_cleaned_pa['age'],users_cleaned_pa['device'])
```

# Convert the relevant variables to a Boolean type

```python
users_cleaned_1= users_cleaned_pa.copy()

users_cleaned_1['default']=users_cleaned_pa['default'].apply(lambda x: 1 if x == "yes" else 0).astype(bool)

users_cleaned_1.head()
```

```python
# Do the same for the other Boolean variables.


var=['housing','loan','term_deposit']
for i in var:
  users_cleaned_1[i]=users_cleaned_pa[i].apply(lambda x: 1 if x == "yes" else 0).astype(bool)
users_cleaned_1.head()
```

```python
users_cleaned_1.info()
```

# Convert `date_joined` to a datetime format

```python
  # Work with a new object.
users_cleaned_2=users_cleaned_1.copy()
users_cleaned_2['date_joined']=pd.to_datetime(users_cleaned_2['date_joined'], format='%Y-%m-%d')

users_cleaned_2['date_joined']
```

```python
users_cleaned_2.info()
```
